```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
import pytrends 
from pytrends.request import TrendReq
```

```python
trends = TrendReq()
```

## Countries where certain keywords are searched the most

```python
trends.build_payload(kw_list=['analytics'])
data = trends.interest_by_region()
data = data.sort_values(by='analytics', ascending=False)
data = data.head(20)
data
```

```python
data.reset_index().plot(x='geoName', y='analytics', figsize=(15,10), kind='bar')
plt.show()
```

```python
data = TrendReq(hl='en_US', tz=360)
data.build_payload(kw_list=['analytics'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15,10))
data['analytics'].plot()
plt.title('Total Google Search Trends over Time', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

## Historical interest overtime

```python
keyword_list = ['python programming']
data = trends.get_historical_interest(keyword_list,year_start=2020,month_start=1, day_start=1, hour_start=0, year_end=2020, month_end=3, day_end=1, hour_end=1)
data = data.sort_values(by='python programming', ascending=False)
data = data.head(30)
data
```

```python
data.reset_index().plot(x='date', y='python programming', figsize=(15,10), kind='bar', color='purple')
plt.show()
```

```python
data = TrendReq(hl='en_US', tz=360)
data.build_payload(kw_list=['python programming'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15,10))
data['python programming'].plot()
plt.title('Total Google Search Trends over Time', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
